# Participant Countries

```python
df_Answer3 = pd.read_sql_query(
    "SELECT AnswerText FROM Answer where QuestionID = 3",
    db_survey
)
df_Answer3.head()
```

```python
Answer3_count = df_Answer3['AnswerText'].value_counts()
Answer3_count 
```

```python
df_Answer3_count = pd.DataFrame(Answer3_count)
df_Answer3_count
```

```python
df_Answer3_count = df_Answer3_count.reset_index()
df_Answer3_count
```

```python
df_Answer3_count = df_Answer3_count.rename(
    columns={"index": "Country", "AnswerText": "Count"}
)
df_Answer3_count
```

```python
df_Answer3_count["Country"] = df_Answer3_count["Country"].replace(
    {
        "-1": "Other",
        "Bahamas, The": "Bahamas",
        "United States of America": 'United States'
    }
)
df_Answer3_count
```
