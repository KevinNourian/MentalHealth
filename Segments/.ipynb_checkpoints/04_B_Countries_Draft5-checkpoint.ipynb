# Countries

```python
df_Answer3 = pd.read_sql_query(
    "SELECT * FROM Answer where QuestionID = 3",
    db_survey
)
df_Answer3.head()
```

```python
df_Answer3["AnswerText"] = df_Answer3["AnswerText"].replace(
    {
        "-1": "Other",
        "Bahamas, The": "Bahamas",
        "United States of America": 'United States'
    }
)
```

```python
df_Answer3 = pd.merge(df_TechinTech, df_Answer3 , on='UserID', how='inner')
df_Answer3 
```

```python
Answer3_count = df_Answer3['AnswerText'].value_counts()
```
